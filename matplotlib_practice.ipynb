```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Create a title.
plt.title("PyBer Fare by Month")

# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Create a title.
plt.title("PyBer Fare by Month")

# Add a grid.
plt.grid()

# Add the legend.
plt.legend()
```

```python
# Annotate the plot using the Object-Oriented Interface

# Create the plot
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)

# 1. Change the line color to green
ax.plot(x_axis, y_axis, color = "green")

# 2. Change the marker to a diamond 
ax.plot(x_axis, y_axis, color = "green", marker = "D")

# 3. Change the line width to 2 points
ax.plot(x_axis, y_axis, color = "green", marker = "D", linewidth = 2)

# 4. Add a legend for the City of Boston
ax.plot(x_axis, y_axis, color = "green", marker = "D", linewidth = 2, label = "Boston")
ax.legend()

# 5. Add a title and axes labels
ax.set_xlabel('Date')
ax.set_ylabel('Fare ($)')
ax.set_title("PyBer Fare by Month")

# 6. Add a y-axis limit
ax.set_ylim(0, 45)

# 7. Add grid lines
ax.grid()
```
